```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, LSTM, GRU
from tensorflow.keras.utils import timeseries_dataset_from_array
```

```python
df = pd.read_csv('./data/clean_data/daily.csv')
df_monthly = pd.read_csv('./data/clean_data/monthly.csv')
df.head()
```

```python
df = pd.get_dummies(data = df, columns = ['store_type', 'location_type', 'region_code'])
```

```python
df.head()
```

```python
store1 = df[df['region_code_R1'] == 1]
```

```python
store1['sales_in_two_months'] = store1['sales'].shift(-60)
store1.dropna(inplace=True)
```

```python
store1.reset_index()
```

```python
# Construct X, y
X = store1.drop(columns=['date', 'sales_in_two_months', 'id', 'store_id'])
y = store1['sales_in_two_months']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle=False)
sc = StandardScaler()
X_train_sc = sc.fit_transform(X_train)
X_test_sc = sc.transform(X_test)
```

```python
sampling_rate = 2
batch_size = 256
seq_len = 30 

train_gen = timeseries_dataset_from_array(
    data=X_train_sc,
    targets=y_train,
    sequence_length=seq_len,
    batch_size=batch_size,
    sampling_rate=sampling_rate
)

val_gen = timeseries_dataset_from_array(
    data=X_test_sc,
    targets=y_test,
    sequence_length=seq_len,
    batch_size=batch_size,
)
```

```python
model = Sequential()
model.add(LSTM(256, return_sequences=True, input_shape=(seq_len, X_train_sc.shape[1])))
model.add(LSTM(128, return_sequences=True))
model.add(LSTM(64))
model.add(Dense(32, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(1))
```

```python
X_train_sc.shape
```

```python
X_test_sc.shape
```

```python
model.compile(loss='mse', optimizer='rmsprop', metrics=['mae'])
```

```python
history = model.fit(train_gen, validation_data=val_gen, epochs=20)
```

```python
plt.plot(history.history['loss'], c='navy', label='Train')
plt.plot(history.history['val_loss'], c='skyblue', label='Val')
plt.legend();
```
